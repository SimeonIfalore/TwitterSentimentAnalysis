```python
# ?creating a function to clean the tweets 
import re
def cleanTxt(text):
    text = re.sub(r'@[A-Za-z0-9]+','', text)  #remove mentions
    text = re.sub(r'#','',text) #removing the '#' symbol
    text = re.sub(r'RT[\s]+', '', text)
    text = re.sub(r'https?:s\/\/S+', '', text) #removing Hyperlinks
    
    return text
    
```

#### Cleaning Gt  bankTweets

```python

GTtweets_df['Text'] = GTtweets_df['Text'].apply(cleanTxt)
```

```python
# further cleaning for Gtbank tweets
def gtclean(text):
    text = re.sub(r'_help +','',text)
    return text
```

```python
GTtweets_df['Text'] = GTtweets_df['Text'].apply(gtclean)
```

```python
# creating function for Subjectivity
def getsubjectivity(text):
    return TextBlob(text).sentiment.subjectivity


# creating the polartity function
def getpolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
# creating the sentiment and polairty DataFrames for Zenith Bank

Zenithtweets_df['subjectivity'] = Zenithtweets_df['Text'].apply(getsubjectivity)


Zenithtweets_df['polarity'] = Zenithtweets_df['Text'].apply(getpolarity)
```

```python
# creating the subjectiivity and polarity columns for the remainder of the banks

# For Access bank 
Accesstweets_df['subjectivity'] = Accesstweets_df['Text'].apply(getsubjectivity)
Accesstweets_df['polarity'] = Accesstweets_df['Text'].apply(getpolarity)


# For GTbank

GTtweets_df['subjectivity'] = GTtweets_df['Text'].apply(getsubjectivity)
GTtweets_df['polarity'] = GTtweets_df['Text'].apply(getpolarity)


# For First bank

Firsttweets_df['subjectivity'] = Firsttweets_df['Text'].apply(getsubjectivity)
Firsttweets_df['polarity'] = Firsttweets_df['Text'].apply(getpolarity)

# For UBA
UBAtweets_df['subjectivity'] = UBAtweets_df['Text'].apply(getsubjectivity)
UBAtweets_df['polarity'] = UBAtweets_df['Text'].apply(getpolarity)
```

```python
# To get the analysis for Negative, Positive and Neutral Tweets let's look at the Analysis using a function

def Analysis(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'
```

```python
# Let's create a new column for our Analysis 

GTtweets_df['analysis'] = GTtweets_df['polarity'].apply(Analysis)

Zenithtweets_df['analysis'] = Zenithtweets_df['polarity'].apply(Analysis)
UBAtweets_df['analysis'] = UBAtweets_df['polarity'].apply(Analysis)
Firsttweets_df['analysis'] = Firsttweets_df['polarity'].apply(Analysis)
Accesstweets_df['analysis'] = Accesstweets_df['polarity'].apply(Analysis)
```
